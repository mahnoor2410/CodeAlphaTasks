```python
import pandas as pd
df = pd.read_csv("car_data.csv")
df.head(3)
```

```python
df.isnull().sum()
```

```python
dummies = pd.get_dummies(df , columns=['Fuel_Type', 'Selling_type', 'Transmission'] , drop_first=True)
dummies=dummies.drop(columns=['Car_Name'])
dummies
```

```python
dummies = dummies.astype(int, errors='ignore')
dummies
```

```python
x = dummies.drop(columns=["Selling_Price"])
x.head()
```

```python
y = dummies["Selling_Price"]
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score

yp = model.predict(X_test)
score_r2 = r2_score(y_test,yp)
score_r2
```
