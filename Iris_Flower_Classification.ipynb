Our goal is to train a model to classify the species based on their measurements (sepal length, sepal width, petal length, petal width).

**solving the problem using Scikit-learn**

```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
dir(iris)
```

```python
iris.data
```

```python
iris.feature_names
```

```python
iris.target
```

```python
iris.target_names
```

```python
df = pd.DataFrame(iris.data , columns=iris.feature_names)
df.head(3)
```

```python
df["target"] = iris.target
df
```

```python
df["flower_name"] = df.target.apply(lambda x: iris.target_names[x]) 
df
```

```python
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.xlabel("sepal length of df0 vs df1")
plt.ylabel("sepal width of df0 vs df1")

plt.scatter(df0["sepal length (cm)"] , df0["sepal width (cm)"] , color = "blue" , marker = "o")
plt.scatter(df1["sepal length (cm)"] , df1["sepal width (cm)"] , color = "red" , marker = "v")
```

```python
plt.xlabel("petal length of df0 vs df1")
plt.ylabel("petal width of df0 vs df1")

plt.scatter(df0["petal length (cm)"] , df0["petal width (cm)"] , color = "blue" , marker = "+")
plt.scatter(df1["petal length (cm)"] , df1["petal width (cm)"] , color = "red" , marker = "*")
```

```python
from sklearn.model_selection import train_test_split
x = df.drop( ["target" , "flower_name"] , axis="columns")
y = df.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y , test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train , y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict( [[4.7,3.2,1.3,0.2]] )
```
